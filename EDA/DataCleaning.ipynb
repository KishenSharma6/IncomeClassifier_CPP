```python
import os
import pandas as pd
import numpy as np
```

```python
#Set notebook preferences
pd.set_option('display.max_columns', 500)
```

```python
#Read-in raw-data
path= "../Data/adultCensusIncome.csv"

raw_df= pd.read_csv(path)
```

**Clean Data**

```python
#Replace ?'s with NaN
os.chdir("../EDA/")
from DataCleaningScripts.DataCleaning import find_missing_values, clean_columns

clean_df= find_missing_values(raw_df)

#Drop unecessary columns
clean_df= clean_columns(clean_df)

#update income
clean_df['income'] = clean_df['income'].apply(lambda x: 1 if x == ">50K" else 0)
```

Dummy Encode Discrete Variables

```python
#Get cols with object data types
discreteCols= list(clean_df.select_dtypes("object").columns)

#Get unique value counts and store in data frame
arr= [0] * len(discreteCols)
for i, col in zip(range(len(discreteCols)), discreteCols):
    arr[i] = clean_df[col].nunique()

    
uniqueCounter= pd.DataFrame(index=discreteCols, data= arr, columns= ["NUnique"])
uniqueCounter
```

```python
#Encode Sex; Male: 1, Female: 0
clean_df['sex']= clean_df['sex'].apply(lambda x: 1 if x == "Male" else 0)

clean_df[:100:10].head()
```

```python
#Remove 'sex' and dummy encode remaining discreteCols
discreteCols.remove("sex")

clean_df= pd.get_dummies(clean_df, columns= discreteCols)
clean_df.shape
```
